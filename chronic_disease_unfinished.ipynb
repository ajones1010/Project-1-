```python
# Dependencies
import pandas as pd
from pathlib import Path
```

```python
#Opening CSV
file = Path('Resources/U.S._Chronic_Disease_Indicators.csv')
```

```python
chronic_disease_df = pd.read_csv(file, encoding="ISO-8859-1")
```

```python
# cleaned_chronic_disease_df = chronic_disease_df [['YearStart', 'YearEnd', 'LocationAbbr', 'LocationDesc', 'DataSource','Topic', 'Question', 'DataValueUnit', 'DataValueType','DataValue', 'DataValueAlt', 'DataValueFootnoteSymbol','DataValueFootnote', 'LowConfideneLimit', 'HighConfidenceLimit','StratificationCategory1', 'Stratification1', 'StratificationCategory2','Stratification2', 'StratificationCategory3', 'Stratification3','Geolocation', 'LocationID', 'TopicID', 'QuestionID', 'ResponseID','DataValueTypeID', 'StratificationCategoryID1', 'StratificationID1']]
chronic_disease_df = chronic_disease_df.fillna(0)
chronic_disease_df.count()
```

```python
# Cleaning df
columns_to_drop = ['Response', 'ResponseID', 'StratificationCategoryID2', 'StratificationID2', 'StratificationCategoryID3', 'StratificationID3', 'Stratification2',	'StratificationCategory3','Stratification3']
columns_existing = [col for col in columns_to_drop if col in chronic_disease_df.columns]

if columns_existing:
    chronic_disease_df = chronic_disease_df.drop(columns=columns_existing)
    print("Columns dropped successfully.")
else:
    print("Columns not found in the DataFrame.")

chronic_disease_df.head()
```
